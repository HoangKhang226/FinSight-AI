## Xem Kết Quả Báo Cáo

```python
import json
import pandas as pd

report_path = "evaluation/ocr/data/custom_finsight/eval_report.json"
try:
    with open(report_path, "r", encoding="utf-8") as f:
        report = json.load(f)

    df = pd.DataFrame.from_dict(report, orient="index")
    print(f"Tổng số file đã hoàn thành: {len(report)}")
    print(f"CER trung bình: {df['cer'].mean():.3f}")
    print(f"WER trung bình: {df['wer'].mean():.3f}")
    
    display(df)
except Exception as e:
    print("Chưa có báo cáo hoặc lỗi:", e)
```
